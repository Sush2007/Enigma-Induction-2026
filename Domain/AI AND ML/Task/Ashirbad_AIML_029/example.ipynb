```python
summarizer = pipeline("summarization", model="t5-small")
```

Enter your paragraph.

```python
text=input("enter the text to be summarized:\n ")
```

```python
word_count = len(text.split())
print("\nWord Count:", word_count)
percent = int(input("Enter summary percentage (e.g. 50 for 50%): "))
```

```python
max_len = int(word_count * percent / 100)
min_len = int(max_len * 0.6)
print("max_len:", max_len)
print("min_len:", min_len)
```

```python
max_len= int(input("enter the maximum length of the summary:\n "))
min_len= int(input("enter the minimum length of the summary:\n "))
print("max_len:", max_len)
print("min_len:", min_len)
```

```python
summary = summarizer(text, max_length=max_len, min_length=min_len, do_sample=False)
if max_len < min_len:
    print("Error: max_length must be greater than or equal to min_length.")
else:
    print("\nSummary:\n")
    print(summary[0]['summary_text'])
### min_length< max_lenghth
```

Yake is an unsupervised keyword extraction algorithm that identifies important terms based on statistical features.

```python
!pip install yake
import yake
```

```python
kw_extractor = yake.KeywordExtractor()
keywords = kw_extractor.extract_keywords(text)
```

```python
print("\n Keywords:")
for keyword in keywords[:6]:
    print(".",keyword)
```

```python
summary_words = len(summary[0]['summary_text'].split())

print("\n Stats:")
print("Original Words:", len(text.split()))
print("Summary Words:", summary_words)
print(f"Reduction: {((len(text.split()) - summary_words)/len(text.split()))*100:.2f}%")
```
