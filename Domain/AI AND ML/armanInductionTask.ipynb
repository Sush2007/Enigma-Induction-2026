Importing pipelines helps us simplify the process of using the model as it handles the tokenisation, math and conversion required. This saves us from the manual workload. Gradio is used for a simple user interface

```python
from transformers import pipeline
import gradio as gr
```

From fetching of the model, function for summariztion to launching the user interface

```python
#to fetch the summarization task model from hugging face, i have chosen bart as it is a good model for summarization
print("The summariaztion model is loading, please wait ...")
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

#Creating a function for the summarization as it makes it easier to use with grdio
def summarize_text(article):
    summary = summarizer(article+ "summarize within 60 words", max_length=60, min_length=20, do_sample=False)
    return summary[0]['summary_text']

# Creation of a simple user interface using Gradio(basic)
demo = gr.Interface(
    fn=summarize_text,
    inputs=gr.Textbox(lines=10, placeholder="Paste the text that is to be summarized here", label="Original Text"),
    outputs=gr.Textbox(lines=4, label="Summary"),
    title="Text Summarizer",
    description="Powered by Meta's BART-large model. Can generate a menaingful and concise summary of the provided text."
    )

# To launch the web application as public
demo.launch(share=True)
```
